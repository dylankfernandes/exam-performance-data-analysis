In this program, we will be using various computational, graphing, and learning libraries to analyze student performance on exams.

First, let's import our data and see what we have to work with. We'll use pandas as our primary method of manipulating our data, numpy as the primary method of running computations on our data, matplotlib and seaborn for graphing our data, and sklearn to analyze the data through regression and classification.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# For those using this tutorial who want to use this notebook on a jupyter notebook that has a dark theme installed.
params = {
    "ytick.color" : "w",
    "xtick.color" : "w",
    "axes.labelcolor" : "w",
    "axes.edgecolor" : "w",
    "text.color": "w"         
}

plt.rcParams.update(params)

df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

So it looks like there are three five primary factors that may contribute towards test scores. Let's separate the men and the women and compare their scores. 

```python
genders = df.groupby('gender')
by_gender = genders[['math score', 'reading score', 'writing score']].mean()
by_gender['total score'] = by_gender['math score'] + by_gender['reading score'] + by_gender['writing score']
by_gender['total score'] = by_gender['total score'] / 3
by_gender.head()
```

So it looks like the females did better overall than the men did. They particularly excelled in reading and writing, but fell behind in math. If we weight the math score and the english score as 50/50 (as seen on standardized tests like the SAT), then we can provide a more accurate depictions.

```python
by_gender['weighted score'] = ((by_gender['reading score'] + by_gender['writing score'])/2 + by_gender['math score'])/2
by_gender.head()
```

Even with this new weighted comparison, the women did better overall than the men did. To further back our claim, let's compare the distributions side by side to check for outliers, shape, center, spread, and any unusual features that we may notice in the data.

```python
bins = [10 * i for i in range(0, 11)]

fig = plt.figure()
fig.subplots_adjust(top=2, right=2, hspace=0.3)

m_math_ax = plt.subplot(321)
m_math_ax.hist(genders.get_group('male')['math score'].values, bins, color='blue')
m_math_ax.set_title('Math Score - Males')

m_writing_ax = plt.subplot(323)
m_writing_ax.hist(genders.get_group('male')['writing score'].values, bins, color='blue')
m_writing_ax.set_title('Writing Score - Males')

m_reading_ax = plt.subplot(325)
m_reading_ax.hist(genders.get_group('male')['reading score'].values, bins, color='blue')
m_reading_ax.set_title('Reading Score - Males')

f_math_ax = plt.subplot(322)
f_math_ax.hist(genders.get_group('female')['math score'].values, bins, color='red')
f_math_ax.set_title('Math Score - Females')

f_writing_ax = plt.subplot(324)
f_writing_ax.hist(genders.get_group('female')['writing score'].values, bins, color='red')
f_writing_ax.set_title('Writing Score - Females')

f_reading_ax = plt.subplot(326)
f_reading_ax.hist(genders.get_group('female')['reading score'].values, bins, color='red')
f_reading_ax.set_title('Reading Score - Females')
```

Great! This gives us a clearer picture of the test scores for males and females. When comparing scores across genders, we can see that the distributions are similar in shape and spread, but the centers for the women are shifted further to the right. This means that outliers didn't play too much of a role in determining the average score, further backing that the women performed better overall for this test.

All of this information begs the question - can we use a regression algorithm to predict test scores based on the given features?

Before we do that, we have to clean up our data. 
* Replace all NaN using the ffill method (replacing the NaN with the data one ahead of it)
* Turn all non-numerical data into numerical data

```python
def convert_to_numeric(x):
    columns = df.columns.values
    for column in columns:
        text_digit_vals = {}
        def convert_to_int(val):
            return text_digit_vals[val]
        if df[column].dtype != np.int64 and df[column].dtype != np.float64:
            column_contents = df[column].values.tolist()
            unique_elements = set(column_contents)
            x = 0
            for unique in unique_elements:
                if unique not in text_digit_vals:
                    text_digit_vals[unique] = x
                    x+=1
            df[column] = list(map(convert_to_int, df[column]))
    return df

df.fillna('ffill', inplace=True)
df = convert_to_numeric(df)
df.head()
```

Now that we have all numeric, non-NaN data, we can run our regression algorithm! To make things simpler on ourselves, we are going to adjust our data so that we only have the total score, so that we don't have three variables to predict.

```python
df['total score'] = (df['math score'] + df['reading score'] + df['writing score']) / 3

X = np.array(df.drop(['total score'], 1))
y = np.array(df['total score'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

clf = linear_model.LinearRegression()
clf.fit(X_train, y_train)
```

Now that we've trained the classifier, it's time to test it out using our testing data.

```python
print(clf.predict(X_test[0:5]), y_test[0:5])
print('Accuracy: ', clf.score(X_test, y_test))
```

Based off of the prediction model, the algorithm predicts the test scores at near perfection!
